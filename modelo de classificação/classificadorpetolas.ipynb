```python
from sklearn import datasets
import numpy as np
```

```python
iris = datasets.load_iris()
```

```python
print(iris)
```

```python
list(iris.keys())
```

```python
#Largura da pétala
X = iris['data'][:,3:]
```

```python
print(X)
```

```python
#se for iris-Virginica, caso contrário, 0 e, agoram será treinado o modelo de regressão lógistica:
y = (iris['target'] == 2).astype(np.int) 
print(y)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_reg = LogisticRegression()
```

```python
log_reg.fit(X,y)
```

```python
# Agora, Calculamos as probabildiades estimadas do modelo para as flores com larguras de pétalas que possuem de 00 a 3 centimentros
```

```python
X_new = np.linspace(0,3,1000).reshape(-1,1)
print(X_new)
```

```python
y_proba = log_reg.predict_proba(X_new)
print(y_proba)
```

```python
import matplotlib.pyplot as plt

# Criar uma figura e um eixo
fig, ax = plt.subplots()

# Plotar os dados com as personalizações
ax.plot(X_new, y_proba[:,1], 'g-', label='iris-Virginica')
ax.plot(X_new, y_proba[:,0], 'b--', label='Não iris-Virginica')
# Adicionar rótulos e título
ax.set_xlabel('Largura da pétalas')
ax.set_ylabel('Probalidade')

# Incluir uma legenda
ax.legend()

# Mostrar o gráfico
plt.show()
```

Na linha pontilhada, temos uma fronteira de decisão que está em torno 1,6 centimentos. Neste ponto, podemos dizer que, com probabildiade de 50%, temos a largura de pétala superior a 1,6cm, e aí teremos o classificador, ind x                                                                                                      xx x x x x x x x x x x x x x x x x x x                                                        icando que é uma iris-virginica ou, ainda, que não seja, caso contrário.
